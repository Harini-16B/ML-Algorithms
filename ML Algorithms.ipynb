There are also many situations where precision and recall are equally important. For example, for our model, if the doctor 
informs us that the patients who were incorrectly classified as suffering from heart disease are equally important since they could be indicative of some other ailment, then we would aim for not only a high recall but a high precision as well.

In such cases, we use something called F1-score. F1-score is the Harmonic mean of the Precision and Recall:    
 F1 Score = 2 * (Precision * Recall) / (Precision + Recall)   


When you cannot afford to miss any detection, you look for high recall; and when you cannot afford to have

#Mean absolute error (MAE):
    It quantifies the average absolute difference between the predicted values and the actual target values in a dataset.
Unlike Mean Squared Error (MSE), which squares the differences, MAE simply takes the absolute value of the differences, 
making it less sensitive to outliers in the data.


The formula for Mean Absolute Error is as follows:

MAE = (1/n) * Σ|yᵢ - ŷᵢ|

Where:

n is the number of data points in the dataset.
yᵢ represents the actual target value for the i-th data point.
ŷᵢ represents the predicted value for the i-th data point

#Mean squared error (MSE):
    It is a common metric used to measure the quality of a regression model's predictions. 
It quantifies the average squared difference between the predicted values and the actual target values in a dataset.
The purpose of squaring the differences is to give more weight to larger errors, which helps penalize significant 
deviations between predictions and actual values.

The formula for Mean Squared Error is as follows:

MSE = (1/n) * Σ(yᵢ - ŷᵢ)²

Where:n is the number of data points in the dataset.
yᵢ represents the actual target value for the i-th data point.
ŷᵢ represents the predicted value for the i-th data point

#Root mean squared error (RMSE):
       RMSE = sqrt((1/n) * Σ(yᵢ - ŷᵢ)²)

#Root mean squared logarithmic error (RMSLE):    
    
 RMSLE = sqrt((1/n)*Σ(log(prediction + 1) - log(actual + 1))^2. 
              RMSLE is then sqrt(mean(squared logarithmic errors))).

Dummy variable -Details in ONE HOT ENCODING

*The Dummy Variable Trap occurs when two or more dummy variables created by one-hot encoding are highly correlated 
(multi-collinear). This means that one variable can be predicted from the others, making it difficult to interpret 
predicted coefficient variables in regression models.

*Dummy variables are qualitative variables or discrete variables that represent categorical data and can take the values
as 0 or 1 to indicate the absence or presence of a specified attribute respectively.

*To avoid dummy variable trap we should always add one less (n-1) dummy variable then the total number of categories present
in the categorical data (n) because the nth dummy variable is redundant as it carries no new information
--Regularization Techniques: Techniques like Ridge or Lasso regression can help mitigate multicollinearity by 
    introducing a penalty term on the coefficients, encouraging the model to use fewer variables and reducing the 
    impact of highly correlated variables.
--Feature Engineering: Instead of using raw categorical variables, you can engineer new features that capture the 
    necessary information without creating perfect multicollinearity.

*Here's what happens if you don't avoid the dummy variable trap:

-Multicollinearity: Including all levels of a categorical variable as dummy variables can create a high degree of correlation 
    among those variables. For example, if you have a categorical variable with three levels (A, B, C) and you create dummy 
    variables D1, D2, and D3, D1 + D2 + D3 will always add up to 1 for every data point. 
    This perfect correlation can lead to multicollinearity, making it difficult for the model to distinguish the individual 
    effects of these dummy variables on the target variable.

-Unreliable Coefficient Estimates: When multicollinearity is present, it becomes challenging to interpret the coefficients 
    of the dummy variables accurately. The coefficients might not make logical sense or can have large standard errors, 
    making them statistically 
-Unstable Model Performance: Multicollinearity can cause the regression model to be sensitive to small changes in the input data.
    This instability can lead to varying coefficient estimates and predictions, making the model less reliable for making 
    predictions on new, unseen data.

-Reduced Model Interpretability: The goal of creating dummy variables is to capture the categorical variable's impact on the
    target variable. However, when multicollinearity is present, it becomes difficult to interpret the effect of 
    each individual dummy variable because their impacts are entangled.

#rfer it-POLYNOMIAL REGRESSION
Polynomial regression is a type of regression analysis used in machine learning to model relationships between independent variables (features) and a dependent variable (target) when the relationship isn't linear. In cases where a simple linear relationship between variables is insufficient to accurately capture the underlying patterns in the data, polynomial regression can be a valuable technique. Here are some reasons why polynomial regression might be needed:

1. **Nonlinear Relationships:** 
    ----------------In many real-world situations, the relationship between variables is not strictly linear.
    Polynomial regression allows for more flexible modeling of nonlinear patterns in the data. 
    
2. **Higher-Degree Patterns:**
    -----------------Polynomial regression can capture patterns that involve higher-degree polynomials 
    (quadratic, cubic, etc.). These patterns might be necessary to accurately describe certain phenomena. Linear regression,
    which only considers linear relationships, might fail to capture the complexities present in the data.

3. **Feature Transformation:** 
    -----------------Polynomial regression can be seen as a way to transform the original features into higher-order 
    features. This transformation can help the model capture interactions between features that have a nonlinear impact on the
    target variable.

4. **Local Behavior:** 
    -----------------In some cases, a relationship might be linear in one range of the data but nonlinear in another range. 
    Polynomial regression can adapt to such local behavior by using different polynomial degrees in different regions of the data.

5. **Underfitting and Overfitting Mitigation:** 
    -----------------While higher-degree polynomial regression models can potentially overfit the
    data, they can also provide a way to mitigate underfitting when the data suggests a more complex relationship. 
    By carefully selecting the polynomial degree and using regularization techniques, you can find a balance that prevents overfitting while capturing the essential nonlinear patterns.

6. **Interpolation:** 
    -----------------Polynomial regression can be useful for interpolating data points between known data points. It can help 
    estimate values between data points based on the available information.

However, it's important to note some considerations and challenges when using polynomial regression:

- **Increased Complexity:** Higher-degree polynomials lead to more complex models with more parameters. This can make the model prone to overfitting, especially when the available data is limited.

- **Choosing the Right Degree:** Selecting the appropriate polynomial degree is crucial. If the degree is too high, the model might capture noise instead of true patterns, leading to poor generalization.

- **Regularization:** Regularization techniques (e.g., Ridge or Lasso) can be applied to polynomial regression to prevent overfitting and control the impact of higher-order terms.

- **Feature Scaling:** When using polynomial features, it's important to ensure that the features are properly scaled to avoid numerical instability.

In summary, polynomial regression is useful when the relationship between variables is nonlinear and requires a more flexible model. However, careful consideration of model complexity, degree selection, and regularization is essential to ensure accurate and robust results.

#Regularization:
Regularization is a critical technique in machine learning that addresses the problem of overfitting. Overfitting occurs 
when a model learns to perform exceptionally well on the training data but fails to generalize effectively to new, unseen data.
Regularization methods help prevent overfitting byadding constraints or penalties to the model's optimization process.
The primary goal of regularization is to encourage the model to learn simpler and more generalizable patterns, leading 
to better performance on unseen data.

#used for solving minimizing the model complexity, penalizing the loss function, reducing the model overfitting(add more bias to 
#reduce model variance, requires a search for optimal penaalty hyperparameter)
#3 main types of regularization:
L1 Regularization (Lasso): Adds a penalty to the absolute values of the model's coefficients, effectively encouraging 
    some coefficients to become exactly zero. This leads to feature selection, where less relevant features are ignored.

L2 Regularization (Ridge): Adds a penalty to the squared values of the model's coefficients. This tends to spread the 
    influence of different features more evenly across the model, reducing the impact of any single feature.

Elastic Net Regularization: Combines both L1 and L2 regularization, striking a balance between feature selection (L1)
    and coefficient shrinkage (L2).

The provided code uses the `StandardScaler` from the `sklearn.preprocessing` module to standardize (scale) the features of a 
dataset. Standardization is a common preprocessing step in machine learning that transforms the features so that they have a 
mean of 0 and a standard deviation of 1. This helps ensure that all features are on a similar scale, 
which can improve the performance of certain algorithms, particularly those that are sensitive to the scale of the features.

Let's break down the code step by step:

1. Importing the Required Library:
   ```python
   from sklearn.preprocessing import StandardScaler
   ```
   This line imports the `StandardScaler` class from the `sklearn.preprocessing` module, which provides the functionality for 
    scaling features.

2. Creating an Instance of StandardScaler:
   ```python
   scaler = StandardScaler()
   ```
   Here, an instance of the `StandardScaler` class is created and stored in the variable `scaler`.

3. Fitting the Scaler to Training Data:
   ```python
   scaler.fit(x_train)
   ```
   The `.fit()` method of the scaler is called with the `x_train` dataset as an argument. This step computes the mean and 
    standard deviation of each feature in the training dataset. These statistics will later be used to transform both the 
    training and testing datasets.

4. Transforming the Training Data:
   ```python
   x_train = scaler.transform(x_train)
   ```
   The `.transform()` method of the scaler is applied to the `x_train` dataset. This step standardizes the features in the 
    training dataset by subtracting the mean and dividing by the standard deviation for each feature. After this operation, 
    the `x_train` dataset will have a mean of approximately 0 and a standard deviation of 1 for each feature.

5. Transforming the Testing Data:
   ```python
   x_test = scaler.transform(x_test)
   ```
   Similarly, the `.transform()` method of the scaler is applied to the `x_test` dataset. The mean and standard deviation values
    obtained during the fitting step with the training data are used to standardize the features in the testing dataset.

By applying the same mean and standard deviation values obtained from the training data, the testing data is scaled in 
the same way as the training data. This is important to ensure that the model is evaluated on data that underwent the same 
preprocessing as the training data.

# KNN - K NEAREST NEIGHBOUR

K-NN algorithm can be used for Regression as well as for Classification but mostly it is used for the Classification problems.
K-NN is a non-parametric algorithm, which means it does not make any assumption on underlying data.
It is also called a lazy learner algorithm because it does not learn from the training set immediately instead it stores the dataset and at the time of classification, it performs an action on the dataset.
KNN algorithm at the training phase just stores the dataset and when it gets new data, then it classifies that data into a category that is much similar to the new data.


The K-NN working can be explained on the basis of the below algorithm:

Step-1: Select the number K of the neighbors
Step-2: Calculate the Euclidean distance of K number of neighbors
Step-3: Take the K nearest neighbors as per the calculated Euclidean distance.
Step-4: Among these k neighbors, count the number of the data points in each category.
Step-5: Assign the new data points to that category for which the number of the neighbor is maximum.
Step-6: Our model is ready.

## Support Vector Machine(SVM)

The goal of the SVM algorithm is to create the best line or decision boundary that can segregate n-dimensional space into classes so that we can easily put the new data point in the correct category in the future. This best decision boundary is called a hyperplane. SVM chooses the extreme points/vectors that help in creating the hyperplane. These extreme cases are called as support vectors, and hence algorithm is termed as Support Vector Machine.

SVMs use kernels to project the data to a higher dimension to seperate the data

#### Separating Hyperplanes

b+b1x1+b2*x2+....bp*xp>0
b0+b1x1+b2*x2+....bp*xp<0

# Tree Based Methods: Decision Tree Learning

A decision tree algorithm is a machine learning algorithm that uses a decision tree to make predictions. It follows a tree-like model of decisions and their possible consequences. The algorithm works by recursively splitting the data into subsets based on the most significant feature at each node of the tree.

A decision tree is a flowchart-like structure in which each internal node represents a test on a feature (e.g. whether a coin flip comes up heads or tails) , each leaf node represents a class label (decision taken after computing all features) and branches represent conjunctions of features that lead to those class.

### Critertion

*Entropy-measure of uncertainity of our data(divide the datasets to minimize the entropy)
*Gini-

### Other Parameters

1. Max_depth:
      maximum depth of the decision tree
2. Min_samples_split:
      Minimum number of samples of node should have before splitting
3. Min_leaf_nodes:
      maximum number of leaves
4. Max_features:   
     Number of feature to consider when looking for the best split           

*easy to interpret
*reqires very large data preparation
*does not support missing values

cost of the tree is log to the number of datapoints used to train the tree
